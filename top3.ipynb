```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
import math
```

```python
data = pd.read_csv('HouseholdPopulationbaseddatasetMIMUTownshipsabbreviated.csv')
```

```python
commu = data[['pcode_st', 'pcode_ts','com_t', 'com_tv','com_radio', 'com_lline', 'com_mob', 'com_comp', 'com_int']]
```

```python
commu.head()
```

```python
X = commu[['com_tv','com_radio', 'com_lline', 'com_mob', 'com_comp', 'com_int']]
X.corr()
```

```python
'''use pca to combine the features into one'''
pca = PCA(n_components = 1)
commu['feature'] = pca.fit_transform(X)
commu
```

```python
a = commu.groupby('pcode_st')['feature', 'com_t'].sum().sort_values()[:6]
a
'''select the lowerest'''
```

```python
b = commu.groupby('pcode_st')['com_t'].sum().sort_values()[:6]
b
'''select the loweest'''
```
